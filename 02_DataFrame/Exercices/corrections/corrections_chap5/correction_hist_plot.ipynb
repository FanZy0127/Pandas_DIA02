```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dataset = {
    'name': ['john', 'mary', 'peter', 'jeff', 'bill', 'lisa', 'jose'],
    'married': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'yes'],
    'age': [23, 78, 22, 19, 45, 33, 20],
    'genre': ['M', 'F', 'M', 'M', 'M', 'F', 'M'],
    'state': ['Paris', 'Lille', 'Paris', 'Lille', 'Paris', 'Bordeaux', 'Bordeaux'],
    'num_children': [3, 0, 2, 4, 3, 1, 5],
    'num_pet': [5, 1, 0, 5, 2, 2, 3]
}

df = pd.DataFrame(dataset)
df
```

```python
# 1. Affichez par état le nombre d'enfant(s).

children_per_state = df.groupby(['state'])['num_children'].sum()

plt.bar(children_per_state.index, children_per_state.values)
plt.show()
```

```python
# 2. Affichez par état le nombre d'enfant(s) des couples mariés.

married_people = df[df['married'] == 'yes']
children_per_married_people_per_state = married_people.groupby(['state'])['num_children'].sum()

plt.bar(children_per_married_people_per_state.index, children_per_married_people_per_state.values)
plt.show()
```

```python
# 3. Est-ce que les personnes mariées ont plus d'enfant(s) ?

children_grouped_by_married = df.groupby(['married'])['num_children'].sum()

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(children_grouped_by_married.index, children_grouped_by_married.values, color='m', width=0.4)
ax.set_ylabel('Number of children')
ax.set_xlabel('Married')
ax.tick_params(axis='x', colors='gray')
ax.tick_params(axis='y', colors='gray')
ax.xaxis.label.set_color('gray')
ax.yaxis.label.set_color('gray')
plt.show()
```

```python
# 4. Est-ce que les personnes divorcées ont plus de chiens que les couples mariés ?

pet_grouped_by_married = df.groupby(['married'])['num_pet'].sum()

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(pet_grouped_by_married.index, pet_grouped_by_married.values, color='mediumseagreen', width=0.4)
ax.set_ylabel('Number of pets')
ax.set_xlabel('Married')
ax.tick_params(axis='x', colors='gray')
ax.tick_params(axis='y', colors='gray')
ax.xaxis.label.set_color('gray')
ax.yaxis.label.set_color('gray')
plt.show()
```
