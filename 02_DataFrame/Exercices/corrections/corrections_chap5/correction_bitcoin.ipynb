```python
import pandas as pd

bitcoins = pd.read_csv('../../data/Cours_BTC2FEUR_-_Feuille_1.csv')
bitcoins.head()
```

```python
# 1. Nettoyer la colonne Prix pour avoir cette colonne dans le type float.

bitcoins['Prix'] = bitcoins['Prix'].str.replace('1 BTC = ', '')
bitcoins['Prix'] = bitcoins['Prix'].str.replace('€', '').astype(float)
```

```python
bitcoins.dtypes
```

```python
bitcoins['Date'] = pd.to_datetime(bitcoins['Date'])
```

```python
bitcoins.dtypes
```

```python
bitcoins.head()
```

```python
# 2. Représentez graphiquement le cours du Bitcoin.

%matplotlib inline
import matplotlib.pyplot as plt

bitcoins.plot(x= 'Date',y ='Prix', rot = 90, figsize = (25,25))
plt.show()
```
