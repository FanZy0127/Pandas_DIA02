```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import randint
```

```python
def play_monty_hall_once(keep_choice) -> bool:
    doors = [0, 1, 2]
    winning_door = randint(0, 2)
    chosen_door = randint(0, 2)
    doors.remove(chosen_door)

    if chosen_door == winning_door:
        doors.remove(doors[randint(0, 1)])
    else:
        doors = [winning_door]

    if not keep_choice:
        chosen_door = doors[0]

    return chosen_door == winning_door


def play_monty_hall(keep_choice, nb_games):
    return [1 if play_monty_hall_once(keep_choice) else 0 for _ in range(nb_games)]


N = 1_000_000

simulations_with_keeping_door = play_monty_hall(True, N)
simulations_without_keeping_door = play_monty_hall(False, N)

count_simulations_with_keeping_door = pd.Series(simulations_with_keeping_door).value_counts()
count_simulations_without_keeping_door = pd.Series(simulations_without_keeping_door).value_counts()

plt.bar([1, 2], [count_simulations_with_keeping_door[1],
                 count_simulations_with_keeping_door[0]],
        tick_label=['Victoires', 'Dfaites'])
plt.title('Monty Hall quand le joueur garde toujours sa porte')
plt.ylabel("Nombre de parties")
plt.show()

plt.bar([1, 2], [count_simulations_without_keeping_door[1],
                 count_simulations_without_keeping_door[0]],
        tick_label=['Victoires', 'Dfaites'])
plt.title('Monty Hall quand le joueur change toujours sa porte')
plt.ylabel("Nombre de parties")
plt.show()
```
